```python
from matplotlib import pyplot as plt
import pandas as pd


graph = pd.read_csv('https://raw.githubusercontent.com/ebubekiraygun/World-Happiness-Data-on-Python/master/2017.csv' )
graph.columns = [c.replace('.', '_') for c in graph.columns]

happy = graph['Happiness_Score']
gdp = graph['Economy__GDP_per_Capita_']


plt.title('Sarah Gailey GDP Per Capita Vs. Happiness Score')

plt.scatter(gdp, happy)
```
